## Analysis of the MetOcean Stokes drift data from the Wadden Sea

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np
import urllib.request

data = json.loads(urllib.request.urlopen('https://plasticadrift.science.uu.nl/galapagosdrifters/waddendrifters_detailed.json').read().decode('utf-8'))
```

Analysis of time between records for each drifter (set at 30 minutes for the first test; 5 minutes for the second test)

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
for i, drifter in enumerate(data.keys()):
    times = np.array([pd.to_datetime(d[0]) for d in data[drifter]], dtype='datetime64[s]')
    dt = np.diff(times)
    I = np.where((dt < np.timedelta64(360, 's')))[0]
    sc = ax.scatter(times[I], (len(I)*[i]), s=40, c=dt[I]/60, vmin=0, vmax=15, cmap='viridis')
    I = np.where((dt > np.timedelta64(360, 's')))[0]
    sc = ax.scatter(times[I], (len(I)*[i]), s=40, c=dt[I]/60, vmin=0, vmax=15, cmap='viridis', edgecolor='k')
ax.set_yticklabels([])
plt.colorbar(sc, ax=ax, label="Difference between consecutive records [minutes]")
ax.set_yticks(range(len(data.keys())), data.keys())
ax.set_ylabel('Drifter name')
ax.set_xlabel('Date of record')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

Analysis of the speed of the drifters as a function of time

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
for i, drifter in enumerate(data.keys()):
    times = np.array([pd.to_datetime(d[0]) for d in data[drifter]], dtype='datetime64[s]')
    lats = [d[1] for d in data[drifter]]
    lons = [d[2] for d in data[drifter]]
    if len(lons) > 0:
        dt = np.diff(times)
        dx = np.diff(lons)*1852*60*np.cos(np.mean(lats)*np.pi/180)
        dy = np.diff(lats)*1852*60
        u = dx/(dt.astype(float))
        v = dy/(dt.astype(float))
        sc = ax.scatter(times[:-1], (len(times)*[i])[:-1], s=40, vmin=0, vmax=1.1, c=np.hypot(u, v), cmap='viridis', marker='s')
ax.set_yticklabels([])
plt.colorbar(sc, ax=ax, label="Drifter speed [m/s]")
ax.set_yticks(range(len(data.keys())), data.keys())
ax.set_ylabel('Drifter name')
ax.set_xlabel('Date of record')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

Analysis of the speed of the drifters as a function of location

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
for i, drifter in enumerate(data.keys()):
    times = np.array([pd.to_datetime(d[0]) for d in data[drifter]], dtype='datetime64[s]')
    lats = [d[1] for d in data[drifter]]
    lons = [d[2] for d in data[drifter]]
    if len(lons) > 0:
        dt = np.diff(times)
        dx = np.diff(lons)*1852*60*np.cos(np.mean(lats)*np.pi/180)
        dy = np.diff(lats)*1852*60
        u = dx/(dt.astype(float))
        v = dy/(dt.astype(float))

        ax.plot(lons, lats, 'k-')
        sc = ax.scatter(lons[:-1], lats[:-1], s=10, c=np.hypot(u, v), label=drifter, zorder=10)

ax.set_aspect('equal')
plt.show()
```

Analysis of the orientation of the drifters (up or down)

```python
fig, ax = plt.subplots(figsize=(15, 5))
for i, drifter in enumerate(data.keys()):
    times = np.array([pd.to_datetime(d[0]) for d in data[drifter]], dtype='datetime64[s]')
    orientation = [d[3] for d in data[drifter]]
    plt.plot(times, [o/2.0 + i - 0.5 for o in orientation], 'k-', lw=1)
plt.yticks(range(len(data.keys())), data.keys())
plt.ylabel('Drifter name')
plt.autoscale(enable=True, axis='x', tight=True)
ax.set_xlabel('Date of record')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

Analysis of battery voltage

```python
fig, ax = plt.subplots(figsize=(15, 5))
colors = plt.cm.jet(np.linspace(0, 1, 24))
linestyles = ['-', '--', '-.', ':']

for i, drifter in enumerate(data.keys()):
    times = np.array([pd.to_datetime(d[0]) for d in data[drifter]], dtype='datetime64[m]')
    voltage = [d[4] for d in data[drifter]]
    ax.plot(times, voltage, label=drifter, color=colors[i], linestyle=linestyles[i%4])
plt.legend(ncol=5)
plt.xlabel('Date of record')
plt.ylabel('Voltage [V]')
plt.show()
```

Analysis of the sea surface temperature at the location of the drifters

```python
fig, ax = plt.subplots(figsize=(15, 5))
for i, drifter in enumerate(data.keys()):
    times = np.array([pd.to_datetime(d[0]) for d in data[drifter]], dtype='datetime64[s]')
    SST = [d[8] for d in data[drifter]]
    plt.scatter(times, (len(SST)*[i]), s=20, c=SST, vmin=8, vmax=12, cmap='viridis', marker='s')
plt.colorbar(label="Sea Surface Temperature [degC]")
plt.yticks(range(len(data.keys())), data.keys())
plt.ylabel('Drifter name')
ax.set_xlabel('Date of record')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

Analysis of sampling intervals

```python
subtract_time_to_fix = False
hist_tinterval_bounds = [270, 330]
fig, ax = plt.subplots(6, 4, figsize=(15, 10))
if(subtract_time_to_fix):
    fig.suptitle('histograms of sample intervals without time to fix in seconds [s], bin size is 1s')
else:
    fig.suptitle('histograms of sample intervals in seconds [s], bin size is 1s')
for i, drifter in enumerate(data.keys()):
    times = np.array([pd.to_datetime(d[0]).timestamp() for d in data[drifter]])
    tinterval = times[1:] - times[:-1]
    if(subtract_time_to_fix):
        times_to_fix = np.array([d[8] for d in data[drifter]])
        tinterval -= times_to_fix[1:]
    ax[int(i/4),i%4].set_title(drifter + ", " + str(len(tinterval)) + " samples")
    ax[int(i/4),i%4].hist(tinterval, bins=range(hist_tinterval_bounds[0], hist_tinterval_bounds[1]+1))
    ax[int(i/4),i%4].set_xlim(hist_tinterval_bounds[0], hist_tinterval_bounds[1])
plt.tight_layout()
plt.show()
```
